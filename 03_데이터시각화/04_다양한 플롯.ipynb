## 1. Bar Chart

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(2021)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")

#barh(), 수평으로 바 생성
plt.barh(y_pos, performance, xerr=error, alpha=0.4)

plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.grid()
plt.show()
```

```python
## 타이타닉 선실별 승객수
```

```python
import seaborn as sns
titanic = sns. load_dataset('titanic')
titanic.head(3)
```

```python
s =titanic['class'].value_counts()
s
```

```python
df = titanic[['class', 'pclass']].groupby('class').count()
df
```

```python
y = s.values
x = np.arange(len(y))
xlabel = s.index.to_list()

plt.title('타이타닉 선실별 승객 수')
plt.bar(x, y)
plt.xticks(x,xlabel)
plt.xlabel('선실')
plt.ylabel('승객수')
plt.show()
```

```python
y = df.pclass.values
x = np.arange(len(y))
xlabel = df.index.to_list()

plt.title('타이타닉 선실별 승객 수')
plt.bar(x, y)
plt.xticks(x,xlabel)
plt.xlabel('선실')
plt.ylabel('승객수',rotation =0, labelpad =15)
plt.show()
```

## 파이 차트
- 카테고리 별 값의 상대저인 비교를 해야 할때

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

## 4. 산점도(Scatter Plot)
- 두개의 실수 데이터 집합의 상관관계를 살펴볼 때 사용

```python
np.random.seed(0)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.title("Scatter Plot")
plt.scatter(X, Y)
plt.show()
```
