```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt


mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family = 'Malgun Gothic')
```

## 예시

```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

### plot 함수
- Pandas의 시리즈나 데이터프레임은 plot이라는 시각화 메서드를 내장하고 있다. - plot은 matplotlib를 내부에서 임포트하여 사용한다.

```python
df1. plot()
plt.title("pandas의 plot() 메소드 사용 예")
plt.xlabel('일자')
plt.ylabel('Data')
plt.grid()
plt.show()
```

## Bar chart

- plot 메서드의 kind라는 인수를 바꾸면 여러가지 플롯을 그릴 수 있다. 다음과 같은 인수값을 지원한다.

    - bar
    - pie
    - hist
    - kde
    - box
    - scatter
    - area

```python

iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
# kind = 'barh'
iris.sepal_length[:20].plot(kind='barh', rot=0)
plt.title("꽃받침의 길이 시각화")
plt.ylabel("Data")
plt.xlabel("꽃받침의 길이")
plt.show()
```

```python
# sort
iris.sepal_length[:20].sort_values().plot(kind='barh', rot=0)
plt.title("꽃받침의 길이 시각화")
plt.ylabel("Data")
plt.xlabel("꽃받침의 길이")
plt.show()
```
