```python
plt.title('라인 플롯')
plt.plot([1,4,9,16]) #y = (x+1)**2
plt.grid()
plt.show()
```

```python
plt.title('라인 플롯')
xs = np.linspace(0,3 ,301)
ys = (xs +1) ** 2
plt.plot(xs, ys) 
plt.grid()
plt.show()
```

## 스타일 지정

```python

# 색깔(r), 마커(s) , 선스타일--
plt.title("'rs--' 스타일의 plot ")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')
plt.grid()
plt.show()
```
